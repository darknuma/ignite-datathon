```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Generate student IDs
student_ids = np.arange(1, 101)

# Generate names (using a placeholder for now)
names = ["Student_" + str(i) for i in student_ids]

# Generate ages (between 15 and 18)
ages = np.random.randint(15, 19, size=100)

# Generate genders (M/F)
genders = np.random.choice(["M", "F"], size=100)

# Generate classes (SS1, SS2, SS3)
classes = np.random.choice(["SS1", "SS2", "SS3"], size=100)

# Generate math scores (between 40 and 100)
math_scores = np.random.randint(40, 101, size=100)

# Generate English scores (between 40 and 100)
english_scores = np.random.randint(40, 101, size=100)

# Generate science scores (between 40 and 100)
science_scores = np.random.randint(40, 101, size=100)

# Generate attendance percentages (between 70 and 100)
attendance = np.random.randint(70, 101, size=100)

# Generate study time (between 2 and 20 hours/week)
study_time = np.random.randint(2, 21, size=100)

# Generate extracurricular activities (a selection of common activities)
extracurricular_activities = np.random.choice(
    [
        "Football",
        "Basketball",
        "Debate Club",
        "Drama Club",
        "Chess Club",
        "Art Club",
        "Science Club",
        "Choir",
        "None",
    ],
    size=100,
)

# Generate learning styles (Visual, Auditory, Kinesthetic, Read/Write)
learning_styles = np.random.choice(
    ["Visual", "Auditory", "Kinesthetic", "Read/Write"], size=100
)

# Create a DataFrame
df = pd.DataFrame(
    {
        "StudentID": student_ids,
        "Name": names,
        "Age": ages,
        "Gender": genders,
        "Class": classes,
        "Math Score": math_scores,
        "English Score": english_scores,
        "Science Score": science_scores,
        "Attendance (%)": attendance,
        "Study Time (hrs/week)": study_time,
        "Extracurricular": extracurricular_activities,
        "Learning Style": learning_styles,
    }
)

# Display the DataFrame
print(df)

# Save the DataFrame to a CSV file
df.to_csv("student_performance_data.csv", index=False)
```
